```python
import pandas as pd
```

```python
stats = pd.read_csv('player_mvp_stats.csv')
del stats["Unnamed: 0"]
stats
```

```python
year = stats.pop("Year")
stats.insert(0, year.name, year)
stats
# stats.shape
```

```python
highest_scoring = stats[stats["G"] > 70].sort_values('PTS', ascending= False).head(20)
highest_scoring
```

```python
highest_scoring.plot.bar("Player", "PTS")
```

```python
highest_scoring = stats.groupby("Year").apply(lambda x: x.sort_values("PTS", ascending=False).head(1))
highest_scoring
```

```python
highest_scoring.plot.bar("Year", 'PTS')
```
