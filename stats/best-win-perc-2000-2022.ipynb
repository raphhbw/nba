```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
teams = pd.read_csv('../teams.csv')
teams
```

```python
teams2000_2022 = teams[teams.Year >= 2000]
teams2000_2022
```

```python
pd.unique(teams2000_2022.Team).shape
```

Fix the issue with team names that have an * at the end

```python
new_teams2000_2022 = teams2000_2022.copy()
new_teams2000_2022.Team = teams2000_2022.Team.apply(lambda t: t[:-1] if (t[-1] == '*') else t)
new_teams2000_2022
```

```python
pd.unique(new_teams2000_2022.Team).shape
```

```python
new_teams2000_2022.groupby(by = 'Team').first()
```

```python
sum_info = new_teams2000_2022.groupby(by = 'Team').sum()
sum_info = sum_info[['W', 'L']]
sum_info
```

```python
data_2000_2022 = sum_info.copy()
data_2000_2022['GP'] =  data_2000_2022.W + data_2000_2022.L
data_2000_2022['W/L%'] = data_2000_2022.W / data_2000_2022.GP
data_2000_2022 = data_2000_2022.sort_values('W/L%')
```

```python
data_2000_2022
```

```python
fig, ax = plt.subplots()
ax.set_title('W/L% per team from 2000-2022')
ax.barh(data_2000_2022[data_2000_2022['W/L%'] < 0.5].index,data_2000_2022[data_2000_2022['W/L%'] < 0.5]['W/L%'], color = 'r', label = 'W/L% < 0.5')
ax.barh(data_2000_2022[data_2000_2022['W/L%'] >= 0.5].index,data_2000_2022[data_2000_2022['W/L%'] >= 0.5]['W/L%'], color = 'b', label = 'W/L% $\geq$ 0.5')
ax.legend()
plt.tight_layout()
# plt.savefig('W%-per-teams-2000-2022.pdf', bbox_inches = 'tight')
plt.show()
```
